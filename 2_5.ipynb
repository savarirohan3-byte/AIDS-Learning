```python
from spacy.language import Language
@Language.component("component_name")
def set_custom_boundaries(doc):
    for token in doc[:-1]:
      if token.text==";":
        doc[token.i+1].is_sent_start=True
    return doc
```

```python
from spacy.tokens import Doc

# Original text
text = '"Management is doing things right; leadership is doing the right things."-Peter Drucker'

# Get tokens and spaces from the nlp object for consistency
doc_tokens = [token.text for token in nlp(text)]
doc_spaces = [token.whitespace_ for token in nlp(text)]

# Create a new Doc object from tokens and spaces
# We pass nlp.vocab to ensure consistency with your loaded model
new_doc = Doc(nlp.vocab, words=doc_tokens, spaces=doc_spaces)

# Manually set is_sent_start for the desired token
# For 'things' which was doc4[4] in the original example
# We need to find its index in doc_tokens
# 'things' is at index 4 (0-indexed) in the token list if we consider the sentence: "Management is doing things right;..."
# Let's verify the index first using the original nlp object for clarity:
original_doc_for_indexing = nlp(text)
print(f"Token at index 4 in original doc: {original_doc_for_indexing[4].text}")

# Now, set is_sent_start for the corresponding token in new_doc
# Assuming 'things' is indeed the 5th token (index 4)
if len(new_doc) > 4:
    new_doc[4].is_sent_start = True

# Now, iterate through the sentences in the new_doc
print('\nSentences in new_doc with custom start at index 4:')
for sent in new_doc.sents:
    print(sent)
```
