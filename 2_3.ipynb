```python
with open(r"reaganomics.txt",encoding='cp1252') as f:
  doc3=nlp(f.read())

```

```python
phrase_list = ['voodoo economics', 'supply-side economics', 'trickle-down economics', 'free-market economics']
phrase_patterns = [nlp(text) for text in phrase_list]
```

```python
converted_patterns = []
for doc_obj in phrase_patterns:
    token_pattern = []
    for token in doc_obj:
        token_pattern.append({"LOWER": token.lower_})
    converted_patterns.append(token_pattern)

matcher.add('VoodooEconomics', converted_patterns)
```

```python
matches=matcher(doc3)
```
